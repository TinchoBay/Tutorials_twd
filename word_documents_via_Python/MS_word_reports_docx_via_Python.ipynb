# Reading data

```python
import pandas as pd
from docx import Document
import pandas as pd
import numpy as npinstall

data_excel = pd.read_csv('titanic.csv')
data_excel.head()
```

```python
data_cols = data_excel.columns.to_list()
data_cols
```

```python
def add_header(document):
    section = document.sections[0]
    header = section.header

    # Add a paragraph to the header with your desired content
    paragraph = header.paragraphs[0]
    paragraph.text = "Titanic - List of passengers"

def add_footer(document):
    section = document.sections[0]
    footer = section.footer

    # Add a paragraph to the footer with your desired content
    paragraph = footer.paragraphs[0]
    paragraph.text = "Report - Passneger Details"
```

```python
for index, row in data_excel.head(3).iterrows():
    document = Document()
    add_header(document)
    add_footer(document)
    document.add_heading('Passenger Details', 0)
    p = document.add_paragraph('The details of the passenger are presented below:')

    table = document.add_table(rows=1, cols=2)
    first_row = table.rows[0].cells
    first_row[0].text = 'Field'
    first_row[1].text = 'Value'
    table.style = 'Table Grid'

    for i in range(len(data_cols)):
        row_cells = table.add_row().cells
        row_cells[0].text = str(data_cols[i])
        row_cells[1].text = str(row[data_cols[i]])

    # export document
    output_file = f'passenger_{row["PassengerId"]}.docx'
    document.save(output_file)
```
